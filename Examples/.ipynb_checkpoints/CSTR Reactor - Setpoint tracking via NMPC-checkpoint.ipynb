```julia
using JuMP
using Ipopt
using LinearAlgebra
using DifferentialEquations#, ForwardDiff, DiffEqSensitivity
using Plots
```

## Simulation setup

```julia
#Model parameters
nc = 4          # [-] number of components
nu = 2          # [-] number of input
k1 = 0.75       # [L/(mol min)] reaction 1 constant
k2 = 1.5       # [L/(mol min)] reaction 2 constant
caIn = 2        # [mol/L] A inlet concentration
cbIn = 1.5      # [mol/L] B inlet concentration
V = 500         # [L] reactor volume
DHr1 = -3.5     # [kcal/mol] reaction 1 enthalpy
DHr2 = -1.5;     # [kcal/mol] reaction 2 enthalpy
```

```julia
#integration parameters
nfe    = 60          # number of control intervals 
ncp    = 3           # number of collocation points
th     = 120         # time horizon
h = th/nfe;           # length of one finite element on the time horizon
```

```julia
#Control parameters
mh = 3 #control horizon
ph = 5 #prediction horizon

Wsp = 100*Matrix{Float64}(I, 2, 2)    # [L^2/mol^2] input usage penalty
Wu = 0.001*Matrix{Float64}(I, 2, 2) # [L^2/mol^2] input usage penalty

QMax = 110      # [kcal/min] maximal heat generation
uMin = [0, 0]      # [L/min] maximum input flowrate
uMax = [50, 50]       # [L/min] maximum input flowrate
DuMax = [10, 10]       # [L/min] maximum input flowrate
        # Concentration C - SP dim: nu X (nfe + 1)
        CSP = [0.44, 0.07] .*ones(1,40)
        CSP = [CSP [0.43,0.12] .*ones(1,21)];
```

```julia
#Plotting setup

# time series for plotting
ts     = Vector{Float64}(undef,nfe) 
for i in 1:nfe
    ts[i] = h*i
end
ts = pushfirst!(ts,0)

# concentration series for plotting
ck = Matrix{Float64}(undef,nc,nfe+1)
# input series for plotting
uk = Matrix{Float64}(undef,nu,nfe+1)
# heat series for plotting
Qk = Vector{Float64}(undef,nfe+1);
```

```julia
# Initial conditions
c0 = [0.5298597648508452,0.06151889952652645,0.4701402351491549,0.10917043266215933]
u0 = [13,13]
Q0 = -V*(k1*c0[1]*c0[2]*DHr1 + k2*c0[2]^2*DHr2);
```

## Function: Formulating reactor discrete dynamics ("plant")

```julia
function ReactorFunction(x0,u)

    #ODE Function
    function f(xdot,x,p,t)

        xdot[1] = -k1*x[1]*x[2] + p[1]*caIn/V - (p[1] + p[2])*x[1]/V
        xdot[2] = -k1*x[1]*x[2] - 2*k2*x[2]^2 + p[2]*cbIn/V - (p[1] + p[2])*x[2]/V
        xdot[3] = k1*x[1]*x[2] - (p[1] + p[2])*x[3]/V
        xdot[4] = k2*x[2]^2 - (p[1] + p[2])*x[4]/V

    end

    # Time t
    t = (0.0,h)

    prob = ODEProblem(f,x0,t,u)
    sol = DifferentialEquations.solve(prob,Vern9(),reltol=1e-8,abstol=1e-8,save_everystep=false)
    xk = sol.u[end]

    Q = -V*(k1*xk[1]*xk[2]*DHr1 + k2*xk[2]^2*DHr2)
    #[-] molar fraction of D
    D = xk[4]/(sum(xk[l] for l in 1:nc))

    return xk, Q, D
end;
```

## Function: Calculating input law using NMPC

```julia
function NMPC(c0,u0,cSP)
    #−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−#
    # Collocation parameters and radau time series
    #N = M^-1 =
    colmat = [0.19681547722366  -0.06553542585020 0.02377097434822;
              0.39442431473909  0.29207341166523 -0.04154875212600;
              0.37640306270047  0.51248582618842 0.11111111111111]
    radau  = [0.15505 0.64495 1.00000]

    #−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−#
    # JuMP model
    m = Model(Ipopt.Optimizer)
    set_silent(m) # avoid printing
    #set_optimizer_attribute(m, "warm_start_init_point", "yes")
    #, warm_start_init_point = "yes", max_iter = 20000, mu_init = 1e-3, replace_bounds = "yes"

    # Set up variables for JuMP
    @variables(m, begin
        c[1:nc, 1:ph, 1:ncp]     #States: [mol/L] concentration (states)
        cdot[1:nc, 1:ph, 1:ncp]  #Differential equations: [mol/L/min]
        u[1:nu, 1:ph]            #Inputs: [L/min]
        du[1:nu, 1:ph]           #Input changes: [L/min]
    end)

    # Set up initial guesses for solver
    for i in 1:ph
        for j in 1:ncp
            set_start_value(c[1,i,j], c0[1])      #C_A
            set_start_value(c[2,i,j], c0[2])    #C_B
            set_start_value(c[3,i,j], c0[3])      #C_C
            set_start_value(c[4,i,j], c0[4])      #C_D
        end

        for mc in 1:nu
            set_start_value(u[mc,i], u0[mc])
            set_start_value(du[mc,i], 0)
        end
    end

    # Set up objective function
    @NLobjective(m, Min, sum(h*(sum(
                                    (
                                    sum(Wsp[kk,kk]*(c[kk + 2,i,j] - cSP[kk])^2 for kk in 1:2)
                                    + sum(Wu[mc,mc]*du[mc,i]^2 for mc in 1:nu)
                                    )*colmat[ncp,j]
                                    for j in 1:ncp
                                   )
                               ) for i in 1:ph
                            )
                )

    @NLexpressions(m, begin
        # [kcal/min] generated heat
        Q[i=1:ph], -V*(k1*c[1,i,ncp]*c[2,i,ncp]*DHr1 + k2*c[2,i,ncp]^2*DHr2)
    end)

    #Set up the constraints
    @NLconstraints(m, begin
        # set up differential equations
        m1[i=1:ph, j=1:ncp], cdot[1,i,j] == -k1*c[1,i,j]*c[2,i,j] + u[1,i]*caIn/V - (u[1,i] + u[2,i])*c[1,i,j]/V
        m2[i=1:ph, j=1:ncp], cdot[2,i,j] == -k1*c[1,i,j] *c[2,i,j]  - 2*k2*c[2,i,j] ^2 + u[2,i]*cbIn/V - (u[1,i] + u[2,i])*c[2,i,j]/V
        m3[i=1:ph, j=1:ncp], cdot[3,i,j] == k1*c[1,i,j] *c[2,i,j]  - (u[1,i] + u[2,i])*c[3,i,j]/V
        m4[i=1:ph, j=1:ncp], cdot[4,i,j] == k2*c[2,i,j] ^2 - (u[1,i] + u[2,i])*c[4,i,j]/V
        # set up collocation equations - 2nd-to-nth point
        coll_c_n[l=1:nc, i=2:ph, j=1:ncp], c[l,i,j] == c[l,i-1,ncp]+h*sum(colmat[j,k]*cdot[l,i,k] for k in 1:ncp)
        # set up collocation equations - 1st point
        coll_c_0[l=1:nc, i=[1], j=1:ncp], c[l,i,j] == c0[l] + h*sum(colmat[j,k]*cdot[l,i,k] for k in 1:ncp)
        #------------- system constraints --------------#
        # set up system constraints (generated heat)
        g1[i=1:ph], Q[i]/QMax - 1 <= 0
        #------------- input constraints --------------#
        # set up du - manipulated variable change
        deltaU_n[mc=1:nu, i=2:ph], du[mc,i] == u[mc,i] - u[mc,i - 1]
        deltaU_0[mc=1:nu, i=[1]], du[mc,i] == u[mc,i] - u0[mc]
        # set up input bounds
        u_UB[mc=1:nu, i=1:ph], u[mc,i] - uMax[mc] <= 0
        u_LB[mc=1:nu, i=1:ph], uMin[mc] - u[mc,i] <= 0
        # set up du bounds
        du_UB[mc=1:nu, i=1:mh], du[mc,i]/DuMax[mc] - 1 <= 0
        du_LB[mc=1:nu, i=1:mh], -1 - du[mc,i]/DuMax[mc] <= 0
        # control horizon - number of MV moves to be optimized at control interval
        deltaU_ph[mc=1:nu, i=mh+1:ph], du[mc,i] == 0
    end)

    #−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−−#
    # Solve the model
    solveNLP = JuMP.optimize!
    status = solveNLP(m)

    # Get values for plotting
    uStar = JuMP.value.(u[:,1]) # time series for plotting

return  uStar

end;
```

# Simulation

```julia
# For plotting
ck[:,1] = c0
uk[:,1] = u0
Qk[1] = Q0

for i in 2:nfe+1
        display("simulation iteration: $i")
        #Optimizing for the interval
        uOpt = NMPC(ck[:,i - 1],uk[:,i - 1],CSP[:,i])

        #integrating the plant
        sol = ReactorFunction(ck[:,i - 1],uOpt)
        ck[:,i] = sol[1]
        Qk[i] = sol[2]
        uk[:,i] = uOpt
end
```
